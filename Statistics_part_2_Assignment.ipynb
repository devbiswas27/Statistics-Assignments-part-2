Q.1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves testing an assumption (called a hypothesis) about a population parameter.

Key Concepts:

Null Hypothesis (H₀):

This is the default or initial assumption. It typically states that there is no effect or no difference.

Example: H₀: The average weight of apples is 150g.

Alternative Hypothesis (H₁ or Ha):

This is what you want to prove. It states that there is an effect or a difference.

Example: H₁: The average weight of apples is not 150g.

Test Statistic:

A value calculated from the sample data that is used to decide whether to reject the null hypothesis.

P-value:

The probability of observing the sample data, or something more extreme, if the null hypothesis is true.

If p-value ≤ significance level (α) (e.g., 0.05), reject H₀.

If p-value > α, fail to reject H₀.

Significance Level (α):

A threshold set by the researcher (commonly 0.05) which determines how much risk of error is acceptable.

Conclusion:

Based on the p-value and the significance level, you either:

Reject the null hypothesis (evidence supports the alternative), or

Fail to reject the null hypothesis (not enough evidence to support the alternative).

Example:
Suppose a factory claims their light bulbs last 1,000 hours on average. You collect a sample and find the average is 950 hours. You perform hypothesis testing to determine whether the difference is statistically significant.

H₀: μ = 1000

H₁: μ ≠ 1000

If your p-value is 0.02 and α = 0.05, you reject H₀ — meaning there's significant evidence the average lifespan is not 1,000 hours.

Q.3.What is the significance level in hypothesis testing, and why is it important?

The significance level, denoted by α (alpha), is the threshold for deciding whether to reject the null hypothesis. It represents the maximum probability of making a Type I error, which is rejecting a true null hypothesis.

 Common Values of α:
0.05 (5%) → Most commonly used

0.01 (1%) → Stricter (used in critical applications)

0.10 (10%) → More lenient (used in exploratory research)

 Why Is It Important?

Controls Risk of False Positives (Type I Errors):
It limits how often you'd incorrectly say something is significant when it's just due to random chance.

Sets the Decision Boundary:

You compare the p-value to α:

If p ≤ α → Reject H₀ (result is statistically significant)

If p > α → Fail to reject H₀ (not enough evidence)

Guides Interpretation of Results:

A lower α means you're being more cautious, while a higher α means you're more willing to take risks.

 Example:

You test whether a new drug lowers blood pressure more than the current one.

H₀: New drug is no better.

H₁: New drug is better.

α = 0.05

Your p-value is 0.03 → p < α, so you reject H₀ and conclude the new drug is significantly better.

Q.4. What does a P-value represent in hypothesis testing?

The p-value (short for probability value) is a key result in hypothesis testing. It measures the probability of observing your sample data—or something more extreme— assuming the null hypothesis is true.

 In Simple Terms:

The p-value tells you how surprising or extreme your sample results are under the assumption that the null hypothesis (H₀) is correct.

 Example:


H₀: A coin is fair (P(Heads) = 0.5)

You flip it 100 times and get 70 heads

The p-value is the probability of getting 70 or more heads just by random chance if the coin is fair

If the p-value is small (e.g., 0.01), that result is very unlikely under H₀, so you may reject H₀.

Q.6.What are Type 1 and Type 2 errors in hypothesis testing?

 >>Type I Error (False Positive)

Definition: Rejecting the null hypothesis (H₀) when it is actually true.

You say there is an effect when there isn’t one.

Probability of this error = α (significance level)

 Example:

You test a new drug and conclude it works better than the current one — but in reality, it doesn’t. That’s a Type I error.

 Type II Error (False Negative)

Definition: Failing to reject the null hypothesis (H₀) when it is actually false.

You say there is no effect, when there really is one.

Probability of this error = β

Power of the test = 1 - β (i.e., the ability to detect an effect when there is one)

 Example:

You test a new drug, but you don’t find a difference, even though the drug actually does work better. That’s a Type II error.

Definition: Tests for an effect in one specific direction.

 Hypotheses Example:

H₀: μ = 75

Or:

H₁ (Left-tailed): μ < 75
(You're only testing for a decline)

 Use a one-tailed test when:

You're only interested in change in one direction.

Example: Is Drug A more effective than Drug B?

 2. Two-Tailed Test

Definition: Tests for an effect in either direction.

You use it when you're checking if a value is simply different, whether it's higher or lower.

 Hypotheses Example:
H₀: μ = 75

H₁: μ ≠ 75
(You’re testing if it’s either greater than or less than 75)

 Use a two-tailed test when:

Any significant difference matters, regardless of direction.

Example: Does a new formula change battery life?

 Key Differences:

Feature	One-Tailed Test	Two-Tailed Test
Alternative (H₁)	μ > μ₀ or μ < μ₀	μ ≠ μ₀
Direction	One direction only	Both directions
Critical region	Entire α in one tail	α is split between two tails

 Example Comparison:

Suppose α = 0.05:

In a one-tailed test, the entire 5% is in one tail of the distribution.

In a two-tailed test, it's split: 2.5% in each tail.

This means a two-tailed test is more conservative (harder to reach significance) because it accounts for surprises in both directions.



Q.8.What is the Z-test, and when is it used in hypothesis testing?

The Z-test is a type of statistical test used in hypothesis testing to determine whether there is a significant difference between sample and population means, or between the means of two samples, under the assumption that the data is normally distributed and the population variance is known.

✅ When to Use a Z-Test:

You use a Z-test in the following conditions:

Large sample size (typically n ≥ 30).

Population variance (σ²) is known or sample standard deviation can approximate it well.

The data is approximately normally distributed, especially important for smaller sample sizes.

You are testing:

A sample mean against a known population mean (one-sample Z-test),

The difference between two means (two-sample Z-test),

A sample proportion vs. a population proportion (Z-test for proportions).

 What the Z-Score Represents:

How unusual your result is, assuming the null is true:

This is compared to a critical value from the standard normal distribution (Z-distribution) to decide whether to reject the null.

Q.13.What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test and T-test are both statistical methods used in hypothesis testing to determine whether there is a significant difference between a sample statistic (usually a mean) and a known or hypothesized population value (or another sample). They are closely related and often used in similar scenarios—but with different assumptions.

 Relationship Between Z-Test and T-Test

    Aspect	                   Z-Test	                                  T-Test
    Purpose	   Test differences in means or proportions	Test differences in means
    Used When	Population standard deviation (σ) is known and/or sample size is large (n ≥ 30)	Population standard deviation is unknown and sample size is small (n < 30)
    Distribution Used	Standard normal distribution (Z-distribution)	Student’s t-distribution
    Shape of Distribution	Fixed (mean = 0, std dev = 1)	Varies depending on degrees of freedom (df)
    Convergence	As sample size increases, t-distribution ≈ Z-distribution	As
    n→∞, T-test ≈ Z-test

Conceptual Connection

Both tests use the same logic:

They compare an observed statistic (like a sample mean) to a value expected under the null hypothesis, standardized by variability.

The main difference lies in the estimate of variability:

Z-test uses the known population standard deviation.

T-test uses the sample standard deviation, which introduces extra uncertainty, especially with small samples. The t-distribution compensates for that with heavier tails.


Q.14.What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.

🎯 What Is a Confidence Interval?

A confidence interval answers this question:

"Given my sample, what range of values is likely to include the true population value?"

It combines:

A point estimate (e.g., sample mean 𝑋
ˉ),

A margin of error (based on variability and confidence level).


 How to Interpret a Confidence Interval

"We are 95% confident that the true population mean lies between X and Y."

This means that if we repeated the study many times and built a 95% CI each time, about 95% of those intervals would contain the true population parameter.

 Important: It does not mean there's a 95% chance that the parameter lies in this specific interval—once the interval is calculated, the true value either is or isn't in it.

Q.20.What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) is a statistical method used to test whether three or more group means are significantly different from each other. Instead of comparing means pair by pair (which increases error), ANOVA uses variance to assess whether at least one group mean differs from the others.

Assumptions of ANOVA:
To ensure valid results, ANOVA relies on several assumptions:

Independence

Observations must be independent of each other.

Usually satisfied by proper random sampling.

Normality

The distribution of the dependent variable should be approximately normal within each group.

Checked using histograms, Q-Q plots, or normality tests (e.g., Shapiro-Wilk).

Homogeneity of Variance (Homoscedasticity)

The variances across groups should be roughly equal.

Tested using Levene’s test or Bartlett’s test.

Q.21.What are the different types of ANOVA tests?

There are several types of ANOVA (Analysis of Variance) tests, each designed for different experimental designs and research questions. All ANOVA tests are used to determine whether group means differ significantly, but they vary based on number of factors, repeated measures, and interactions.

🔹 1. One-Way ANOVA

Purpose: Compare means across three or more independent groups using one independent variable (factor).

Example: Compare average test scores across 3 teaching methods.

Assumptions:

One factor

Independent samples

Normal distribution

Equal variances across groups

🔹 2. Two-Way ANOVA

Purpose: Assess the effect of two independent variables (factors), and whether there is an interaction between them.

Example: Examine how teaching method and school type affect test scores.

Outputs:

Main effect of Factor A

Main effect of Factor B

Interaction effect (A × B)

🔹 3. Repeated Measures ANOVA

Purpose: Compare means within the same group over multiple conditions or time points.

Example: Measure blood pressure before, during, and after treatment on the same patients.

Key Point:

Accounts for correlation between repeated observations

🔹 4. Mixed-Design ANOVA (Split-Plot ANOVA)

Purpose: Combine between-subjects and within-subjects factors.

Example: Compare two drug treatments (between-subjects) across 3 time points (within-subjects).

Use when:

You have one repeated measure and one independent group variable

🔹 5. MANOVA (Multivariate ANOVA)

Purpose: Test differences in multiple dependent variables simultaneously.

Example: Compare how diet affects weight, cholesterol, and blood pressure.

Why use it?

Controls for Type I error across multiple outcomes

Accounts for relationships between dependent variables

```python
#Q.2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
low = 1    # inclusive
high = 7   # exclusive (so values will be 1 through 6)

# Create the distribution
dist = randint(low, high)

# Values over the support
x = np.arange(low, high)

# Get the PMF values
pmf = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf, use_line_collection=True)  # Corrected here
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


```

```python
#Q.3.> Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    - x : int (0 or 1) — the outcome
    - p : float — probability of success (0 <= p <= 1)

    Returns:
    - Probability of observing x under Bernoulli(p)
    """
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print(bernoulli_pdf(1, 0.7))  # Probability of success
print(bernoulli_pdf(0, 0.7))  # Probability of failure
```

```python
#Q.4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success in each trial
size = 1000  # number of samples to simulate

# Simulate binomial distribution samples
samples = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(samples, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='skyblue', edgecolor='black')

plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q.9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    - data: list or numpy array of numerical values

    Returns:
    - z_scores: numpy array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Avoid division by zero
    if std == 0:
        raise ValueError("Standard deviation is zero, Z-scores cannot be calculated.")

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.stem(range(len(z_scores)), z_scores, use_line_collection=True)
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

# Example usage:
data = [10, 12, 23, 23, 16, 23, 21, 16]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """Calculate Z-score for a given data point."""
    if std == 0:
        raise ValueError("Standard deviation cannot be zero.")
    return (x - mean) / std

def compare_and_plot(z):
    """Compare Z-score to standard normal and plot."""
    # Calculate cumulative probability and p-value
    cdf = norm.cdf(z)
    p_value = 2 * (1 - cdf) if z > 0 else 2 * cdf

    print(f"Z-score: {z:.4f}")
    print(f"Cumulative Probability (P(Z <= z)): {cdf:.4f}")
    print(f"Two-tailed p-value: {p_value:.4f}")

    # Plot standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')

    # Fill area under the curve
    x_fill = np.linspace(-4, z, 1000) if z >= 0 else np.linspace(z, 4, 1000)
    plt.fill_between(x_fill, norm.pdf(x_fill), color='skyblue', alpha=0.5, label='Shaded Area')

    # Add vertical line at Z-score
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    # Labels and legend
    plt.title('Standard Normal Distribution and Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
x = 75
mean = 70
std = 5

z = calculate_z_score(x, mean, std)
compare_and_plot(z)



```

```python
#Q15.Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample: list or numpy array of sample data
    - mu_0: hypothesized population mean (null hypothesis)
    - sigma: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z-statistic
    - p_value: two-tailed p-value
    - decision: whether to reject the null hypothesis
    """
    sample = np.array(sample)
    n = len(sample)
    sample_mean = np.mean(sample)

    # Z-statistic
    z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Hypothesis decision
    reject = p_value < alpha
    decision = "Reject null hypothesis" if reject else "Fail to reject null hypothesis"

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    return z_stat, p_value, decision

# 🧾 Example Usage:
sample_data = [71, 69, 73, 68, 70, 72, 70, 69, 74, 71]  # sample of size 10
population_mean = 70
population_std = 2  # known σ

z_test(sample_data, population_mean, population_std)

```

```python
#Q.16.  Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the population mean using the t-distribution.

    Parameters:
    - data: list or numpy array of sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - (lower_bound, upper_bound): the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error = s / sqrt(n)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.4f}, {upper:.4f})")
    return (lower, upper)

# 🧾 Example Usage
data = [71, 69, 73, 68, 70, 72, 70, 69, 74, 71]
confidence_interval(data)
```

```python
#Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    std_error = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Population Std Dev = {pop_std}")
    print(f"Sample Size = {n}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. Sample mean is significantly different from population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. No significant difference detected.")


```

```python
#Q3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Standard error of the mean
    std_error = pop_std / np.sqrt(n)

    # Z statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Population Std Dev = {pop_std}")
    print(f"Sample Size = {n}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Decision
    if p_value < alpha:
        print(f"Reject the null hypothesis at α={alpha}. Sample mean is significantly different.")
    else:
        print(f"Fail to reject the null hypothesis at α={alpha}. No significant difference detected.")

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=105, scale=15, size=40)  # Sample from population with mean ~105

one_sample_z_test(sample_data, pop_mean=100, pop_std=15)
```

```python
#Q4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Critical z-values for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)
    print(f"Critical Z-values: ±{z_critical:.4f}")

    # Decision
    if abs(z_stat) > z_critical:
        print(f"Reject the null hypothesis at α={alpha}.")
    else:
        print(f"Fail to reject the null hypothesis at α={alpha}.")

    # Plot the standard normal distribution and critical regions
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

    # Plot the Z statistic
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')

    plt.title('Two-tailed Z-test: Standard Normal Distribution')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
np.random.seed(123)
sample_data = np.random.normal(loc=103, scale=15, size=50)  # Sample mean around 103
two_tailed_z_test(sample_data, pop_mean=100, pop_std=15, alpha=0.05)
```

```python
#Q5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, pop_std, sample_size, alt_mean, alpha=0.05):
    # Calculate critical Z value for alpha
    z_crit = norm.ppf(1 - alpha)

    # Standard error
    se = pop_std / np.sqrt(sample_size)

    # Critical value in original units
    crit_val = pop_mean + z_crit * se

    # Type II error calculation (beta)
    # Beta is probability that sample mean falls below critical value given H_a true
    beta = norm.cdf(crit_val, loc=alt_mean, scale=se)

    print(f"Significance level (Type I error, α): {alpha}")
    print(f"Critical value (threshold): {crit_val:.4f}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Power of test (1 - β): {1 - beta:.4f}")

    # Plot distributions
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 1000)
    null_dist = norm.pdf(x, loc=pop_mean, scale=se)
    alt_dist = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label='Null Hypothesis Distribution', color='blue')
    plt.plot(x, alt_dist, label='Alternative Hypothesis Distribution', color='green')

    # Shade Type I error (α) region: right tail of null distribution beyond critical value
    x_fill1 = np.linspace(crit_val, x[-1], 200)
    plt.fill_between(x_fill1, 0, norm.pdf(x_fill1, loc=pop_mean, scale=se), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error (β) region: left tail of alt distribution below critical value
    x_fill2 = np.linspace(x[0], crit_val, 200)
    plt.fill_between(x_fill2, 0, norm.pdf(x_fill2, loc=alt_mean, scale=se), color='orange', alpha=0.3, label='Type II Error (β)')

    # Draw vertical line at critical value
    plt.axvline(crit_val, color='black', linestyle='--', label=f'Critical Value = {crit_val:.2f}')

    # Annotate means
    plt.annotate('μ₀ (H₀ mean)', xy=(pop_mean, norm.pdf(pop_mean, pop_mean, se)), xytext=(pop_mean-1, 0.05),
                 arrowprops=dict(facecolor='blue', shrink=0.05), color='blue')
    plt.annotate('μ₁ (Hₐ mean)', xy=(alt_mean, norm.pdf(alt_mean, alt_mean, se)), xytext=(alt_mean+1, 0.05),
                 arrowprops=dict(facecolor='green', shrink=0.05), color='green')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage:
plot_type1_type2_errors(pop_mean=100, pop_std=15, sample_size=30, alt_mean=110, alpha=0.05)
```

```python
#Q6.Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    # Perform independent two-sample t-test
    t_stat, p_value = ttest_ind(sample1, sample2)

    print(f"Sample 1 mean: {np.mean(sample1):.4f}, n = {len(sample1)}")
    print(f"Sample 2 mean: {np.mean(sample2):.4f}, n = {len(sample2)}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. The means are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. No significant difference between the means.")

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=50, scale=5, size=30)  # group 1 data
group2 = np.random.normal(loc=52, scale=5, size=30)  # group 2 data

independent_t_test(group1, group2)
```

```python
#Q7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(before, after, alpha=0.05):
    # Perform paired t-test
    t_stat, p_value = ttest_rel(before, after)

    print(f"Mean Before: {np.mean(before):.4f}")
    print(f"Mean After: {np.mean(after):.4f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. Significant difference detected.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. No significant difference detected.")

    # Visualization
    plt.figure(figsize=(10,6))

    # Boxplots
    plt.boxplot([before, after], labels=['Before', 'After'])

    # Connect paired points
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], 'gray', marker='o', alpha=0.5)

    plt.title('Paired Sample Comparison')
    plt.ylabel('Values')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
np.random.seed(42)
before_treatment = np.random.normal(loc=70, scale=10, size=20)
after_treatment = before_treatment + np.random.normal(loc=-5, scale=8, size=20)  # simulated improvement

paired_t_test(before_treatment, after_treatment)
```

```python
#Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import norm, t

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    return z_stat, p_value

def one_sample_t_test(sample, pop_mean, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    se = sample_std / np.sqrt(n)
    t_stat = (sample_mean - pop_mean) / se
    p_value = 2 * (1 - t.cdf(abs(t_stat), df=n-1))
    return t_stat, p_value

# Simulate data
np.random.seed(123)
population_mean = 100
population_std = 15  # known for Z-test
sample_size = 30

sample_data = np.random.normal(loc=102, scale=15, size=sample_size)

# Perform tests
z_stat, z_p = one_sample_z_test(sample_data, population_mean, population_std)
t_stat, t_p = one_sample_t_test(sample_data, population_mean)

print("One-sample Z-test:")
print(f"  Z-statistic = {z_stat:.4f}, P-value = {z_p:.4f}")

print("\nOne-sample T-test:")
print(f"  T-statistic = {t_stat:.4f}, P-value = {t_p:.4f}")

# Interpretation
alpha = 0.05
print("\nInterpretation:")
if z_p < alpha:
    print("  Z-test: Reject null hypothesis - significant difference.")
else:
    print("  Z-test: Fail to reject null hypothesis - no significant difference.")

if t_p < alpha:
    print("  T-test: Reject null hypothesis - significant difference.")
else:
    print("  T-test: Fail to reject null hypothesis - no significant difference.")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=30)

ci_low, ci_up = confidence_interval(data, confidence=0.95)
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_up:.2f})")
```

```python
#Q10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_dev = np.std(sample, ddof=1)  # sample std deviation
    std_err = std_dev / np.sqrt(n)

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_error = t_crit * std_err
    return margin_error

# Example usage:
np.random.seed(1)
data = np.random.normal(loc=100, scale=15, size=40)

me = margin_of_error(data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {me:.4f}")
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):
    # Total probability of evidence
    P_E = likelihood_E_given_H * prior_H + likelihood_E_given_not_H * prior_not_H

    # Posterior probability P(H|E)
    posterior = (likelihood_E_given_H * prior_H) / P_E
    return posterior

# Given values
prior_disease = 0.01                # P(H) - disease prevalence
prior_no_disease = 1 - prior_disease

likelihood_positive_given_disease = 0.99   # P(E|H) - true positive rate
likelihood_positive_given_no_disease = 0.01 # P(E|~H) - false positive rate

posterior = bayes_theorem(
    prior_disease,
    likelihood_positive_given_disease,
    prior_no_disease,
    likelihood_positive_given_no_disease
)

print(f"Posterior probability of having disease given positive test: {posterior:.4f}")
```

```python
#Q12.Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
from scipy.stats import chi2_contingency

# Example data: Contingency table of two categorical variables
# Rows = categories of variable A, Columns = categories of variable B
observed = np.array([
    [30, 10, 20],
    [20, 20, 15],
    [10, 20, 25]
])

chi2_stat, p_value, dof, expected = chi2_contingency(observed)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variables are dependent (associated).")
else:
    print("Fail to reject the null hypothesis: Variables are independent.")
```

```python
#Q13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed):
    row_sums = observed.sum(axis=1, keepdims=True)  # sum of each row, shape (rows, 1)
    col_sums = observed.sum(axis=0, keepdims=True)  # sum of each column, shape (1, cols)
    grand_total = observed.sum()

    expected = (row_sums @ col_sums) / grand_total  # matrix multiplication to get expected counts
    return expected

# Example observed data (contingency table)
observed = np.array([
    [30, 10, 20],
    [20, 20, 15],
    [10, 20, 25]
])

expected_freq = calculate_expected_frequencies(observed)
print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected_freq)
```

```python
#Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Observed data counts (example)
observed = np.array([50, 30, 20])

# Expected proportions for each category (must sum to 1)
expected_probs = np.array([0.4, 0.4, 0.2])

# Convert expected proportions to expected counts
expected = expected_probs * observed.sum()

# Perform goodness-of-fit Chi-square test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data do NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fit the expected distribution.")
```
